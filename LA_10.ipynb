```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully!\n")

print("--- Features and their Data Types ---")
df.info()

print("\n--- Summary of Feature Types ---")
print("Numeric (Continuous): sepal_length, sepal_width, petal_length, petal_width")
print("Nominal (Categorical): species")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Histograms of Iris Features', fontsize=16)

sns.histplot(df['sepal_length'], kde=True, ax=axes[0, 0], color='skyblue')
sns.histplot(df['sepal_width'], kde=True, ax=axes[0, 1], color='olive')
sns.histplot(df['petal_length'], kde=True, ax=axes[1, 0], color='gold')
sns.histplot(df['petal_width'], kde=True, ax=axes[1, 1], color='teal')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=df.drop('species', axis=1), palette='Set2')

plt.title('Box Plots of Iris Features (Detecting Outliers)', fontsize=16)
plt.ylabel('Measurements (cm)')
plt.show()
```
