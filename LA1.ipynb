```python
import pandas as pd
import numpy as np

print("Libraries imported successfully!")
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)

print("Iris Data loaded successfully! Here are the first 5 rows:")
display(df.head())
```

```python
print("Dimensions of the dataset (Rows, Columns):", df.shape)
print("-" * 50)

print("Information about variables and their data types:")
df.info()
print("-" * 50)

print("Count of missing values in each column:")
print(df.isnull().sum())
print("-" * 50)

print("Initial statistics for numerical columns:")
display(df.describe())
```

```python
print("Current Data Types:\n", df.dtypes)
print("-" * 50)

df['species'] = df['species'].astype('category')

print("Data Types after conversion:\n", df.dtypes)
```

```python
df_encoded = pd.get_dummies(df, columns=['species'])

df_encoded['species_setosa'] = df_encoded['species_setosa'].astype(int)
df_encoded['species_versicolor'] = df_encoded['species_versicolor'].astype(int)
df_encoded['species_virginica'] = df_encoded['species_virginica'].astype(int)

print("Dataset after converting the categorical 'species' variable into quantitative variables:")
display(df_encoded.head())
display(df_encoded.tail())
```
