```python
import nltk
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Download required NLTK data packs (Updated for newer NLTK versions)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng') # <-- Updated line
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

print("All NLTK packages downloaded successfully!")
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "The quick brown foxes are jumping over the lazy dogs. Machine learning is fascinating!"
print("Original Document:\n", text, "\n")

tokens = word_tokenize(text)
print("--- 1. Tokenization ---")
print(tokens)
print("\n")

pos_tags = pos_tag(tokens)
print("--- 2. POS Tagging ---")
print(pos_tags)
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
filtered_tokens = [word for word in filtered_tokens if word.isalnum()]

print("--- 3. Stop Words Removal ---")
print(filtered_tokens)
print("\n")

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("--- 4. Stemming ---")
print(stemmed_words)
print("\n")

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("--- 5. Lemmatization ---")
print(lemmatized_words)
```

```python
documents = [
    "The quick brown foxes are jumping.",
    "Machine learning is fascinating.",
    "The brown foxes are learning."
]

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

feature_names = tfidf_vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("--- TF-IDF Representation ---")
pd.set_option('display.max_columns', None)
display(df_tfidf.round(4))
```
