```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully!")
print("Initial Dimensions:", df.shape)
display(df.head())
```

```python
df.loc[0:5, 'math score'] = np.nan
df.loc[10:15, 'lunch'] = np.nan

print("--- Missing Values Before ---")
print(df.isnull().sum())
print("-" * 50)

df['math score'].fillna(df['math score'].median(), inplace=True)

df['lunch'].fillna(df['lunch'].mode()[0], inplace=True)

print("--- Missing Values After ---")
print(df.isnull().sum())
```

```python
def treat_outliers_iqr(dataframe, column_name):
    Q1 = dataframe[column_name].quantile(0.25)
    Q3 = dataframe[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = dataframe[(dataframe[column_name] < lower_bound) | (dataframe[column_name] > upper_bound)]
    print(f"{column_name}: Found {len(outliers)} outliers. Capping them...")

    dataframe[column_name] = np.where(dataframe[column_name] > upper_bound, upper_bound, dataframe[column_name])
    dataframe[column_name] = np.where(dataframe[column_name] < lower_bound, lower_bound, dataframe[column_name])

columns_to_check = ['math score', 'reading score', 'writing score']

for col in columns_to_check:
    treat_outliers_iqr(df, col)

print("\nOutlier treatment complete.")
```

```python
scaler = MinMaxScaler()
df['math_scaled'] = scaler.fit_transform(df[['math score']])
print("1. Applied Min-Max Scaling to 'math score'.")

df['reading_sqrt'] = np.sqrt(df['reading score'])
print("2. Applied Square Root Transformation to 'reading score'.")

df['writing_log'] = np.log1p(df['writing score'])
print("3. Applied Log Transformation to 'writing score'.")

print("\n--- Transformed Variables Sample ---")
display(df[['math_scaled', 'reading_sqrt', 'writing_log']].head())
```
