```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset Loaded Successfully!\n")

summary_stats = df.groupby('species')['sepal_length'].agg(['mean', 'median', 'min', 'max', 'std'])

print("--- Summary Statistics of Sepal Length grouped by Species ---")
display(summary_stats)
print("\n")

species_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
numeric_species_list = df['species'].map(species_mapping).tolist()

print("--- Numeric List representing the Categorical Variable ---")
print(f"Total items in list: {len(numeric_species_list)}")
print("First 15 items in the list (representing the first 15 flowers, which are all setosa):")
print(numeric_species_list[:15])
```

```python
statistical_details = df.groupby('species').describe()

print("--- Basic Statistical Details for Iris-setosa ---")
display(statistical_details.loc['setosa'])

print("\n--- Basic Statistical Details for Iris-versicolor ---")
display(statistical_details.loc['versicolor'])

print("\n--- Basic Statistical Details for Iris-virginica ---")
display(statistical_details.loc['virginica'])
```
