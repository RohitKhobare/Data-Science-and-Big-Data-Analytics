```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic_df = sns.load_dataset('titanic')

print("Dataset loaded successfully!")
print(f"Dimensions: {titanic_df.shape[0]} rows and {titanic_df.shape[1]} columns.")
display(titanic_df.head())
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(data=titanic_df, x='class', y='survived', hue='sex', palette='muted', errorbar=None)

plt.title('Survival Rate by Passenger Class and Gender')
plt.ylabel('Survival Probability')
plt.xlabel('Passenger Class')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=titanic_df, x='fare', bins=40, kde=True, color='teal')

plt.title('Distribution of Titanic Ticket Prices (Fare)')
plt.xlabel('Fare Price')
plt.ylabel('Number of Passengers')

plt.show()
```
