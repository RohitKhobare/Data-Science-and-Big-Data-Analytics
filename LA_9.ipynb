```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = sns.load_dataset('titanic')

plt.figure(figsize=(10, 6))

sns.boxplot(data=titanic_df, x='sex', y='age', hue='survived', palette='Set2')

plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender (Sex)')
plt.ylabel('Age (in years)')

plt.legend(title='Survived', loc='upper right', labels=['No (0)', 'Yes (1)'])

plt.show()
```
